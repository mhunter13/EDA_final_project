```python
import pandas as pd
```

```python
stations = pd.read_csv("NYC_Transit_Subway_Entrance_And_Exit_Data.csv")
stations.head()
```

```python
stations.columns
```

```python
stations.drop(stations.columns[5:], axis = 1, inplace=True)
stations.head()
```

```python
stations.groupby('Station Name').count()
```

```python
#why does Yankee Stadium 161st St have 16 different entries?
mask = stations['Station Name'] == 'Yankee Stadium-161st St'
stations[mask]
```

```python
stations.drop_duplicates(inplace=True)
stations.info()
```

```python
df_stations = stations.groupby('Station Name').mean().reset_index()
df_stations = df_stations.round(6)  
df_stations.head()
```

```python
df_stations.info()
```

```python
df_stations.columns
```

```python
# to_replace = ['rd', 'th', 'nd','st','']
# df_stations['Station Name'] = df_stations['Station Name'].replace(to_replace,'')
df_stations['Station Name'] = df_stations['Station Name'].str.replace('th','')
df_stations['Station Name'] = df_stations['Station Name'].str.replace('nd','')
df_stations['Station Name'] = df_stations['Station Name'].str.replace('rd','')
df_stations['Station Name'] = df_stations['Station Name'].str.replace('st','')
df_stations = df_stations.sort_values('Station Name').reset_index()
df_stations.head(50)
```

```python
df_stations.iloc[50:100]
```

```python
((df_stations.groupby('Station Name').count() > 1) == True).count()
```

```python
df_stations.to_csv('station_list.csv')
```
